```python
import pandas as pd 
```

```python
file_path_bat =  r'C:\Users\sarth\OneDrive\Documents\Mobiqliq\Data processing\IPL\IPLT20-BattingBowlingDetails\Chennai Super Kings-BattingDetails.csv'
df_bat = pd.read_csv(file_path_bat)

df_bat['30s'] = df_bat['runs'].apply(lambda x: 1 if 30 <= x < 50 else 0)
df_bat['50s'] = df_bat['runs'].apply(lambda x: 1 if 50 <= x < 100 else 0)
df_bat['100s'] = df_bat['runs'].apply(lambda x: 1 if x >= 100 else 0)
df_bat['ducks'] = df_bat['runs'].apply(lambda x: 1 if x == 0 else 0)
df_bat.head()
```

```python
file_path_bowl = r'C:\Users\sarth\OneDrive\Documents\Mobiqliq\Data processing\IPL\IPLT20-BattingBowlingDetails\Chennai Super Kings-BowlingDetails.csv'
df_bowl = pd.read_csv(file_path_bowl)
df_bowl['3 wicket'] = df_bowl['wicket'].apply(lambda x: 1 if x==3 else 0)
df_bowl['4 wicket'] = df_bowl['wicket'].apply(lambda x: 1 if x==4 else 0)
df_bowl['5 wicket'] = df_bowl['wicket'].apply(lambda x: 1 if x==5 else 0)
df_bowl.head()
```

```python
df_bat['dr11score'] = df_bat['runs'] + df_bat['4s'] + df_bat['6s'] + (df_bat['30s']*4) + (df_bat['50s']*8) + (df_bat['100s']*16) - (df_bat['ducks']*2)
df_bat.head()
```

```python
df_bowl['dr11score'] = (df_bowl['wicket']*25) + (df_bowl['3 wicket']*4) + (df_bowl['4 wicket']*8) + (df_bowl['5 wicket']*16) + (df_bowl['maidens']*12)
df_bowl.head()
```

```python
file_path_bat_1 =  r'C:\Users\sarth\OneDrive\Documents\Mobiqliq\Data processing\IPL\IPLT20-BattingBowlingDetails\Delhi Daredevils-BattingDetails.csv'
df_bat_1 = pd.read_csv(file_path_bat_1)

df_bat_1['30s'] = df_bat_1['runs'].apply(lambda x: 1 if 30 <= x < 50 else 0)
df_bat_1['50s'] = df_bat_1['runs'].apply(lambda x: 1 if 50 <= x < 100 else 0)
df_bat_1['100s'] = df_bat_1['runs'].apply(lambda x: 1 if x >= 100 else 0)
df_bat_1['ducks'] = df_bat_1['runs'].apply(lambda x: 1 if x == 0 else 0)
df_bat_1['dr11score'] = df_bat_1['runs'] + df_bat_1['4s'] + df_bat_1['6s'] + (df_bat_1['30s']*4) + (df_bat_1['50s']*8) + (df_bat_1['100s']*16) - (df_bat_1['ducks']*2)
df_bat_1.head()
```

```python
file_path_bowl_1 = r'C:\Users\sarth\OneDrive\Documents\Mobiqliq\Data processing\IPL\IPLT20-BattingBowlingDetails\Delhi Daredevils-BowlingDetails.csv'
df_bowl_1 = pd.read_csv(file_path_bowl_1)
df_bowl_1['3 wicket'] = df_bowl_1['wicket'].apply(lambda x: 1 if x==3 else 0)
df_bowl_1['4 wicket'] = df_bowl_1['wicket'].apply(lambda x: 1 if x==4 else 0)
df_bowl_1['5 wicket'] = df_bowl_1['wicket'].apply(lambda x: 1 if x==5 else 0)
df_bowl_1['dr11score'] = (df_bowl_1['wicket']*25) + (df_bowl_1['3 wicket']*4) + (df_bowl_1['4 wicket']*8) + (df_bowl_1['5 wicket']*16) + (df_bowl_1['maidens']*12)
df_bowl_1.head()
```

```python
bat = pd.concat([df_bat, df_bat_1], ignore_index=True)
```

```python
# Select only the numeric columns
numeric_cols = bowl.select_dtypes(include=[np.number]).columns

# Check for infinite values in the numeric columns
inf_columns = bowl[numeric_cols].columns.to_series()[np.isinf(bowl[numeric_cols]).any()]
print("Columns with infinite values:", inf_columns)
```

```python
# Replace inf and -inf with NaN in 'econrate' column in the 'bowl' DataFrame
bowl['econrate'] = bowl['econrate'].replace([np.inf, -np.inf], np.nan)
# Fill NaN values in 'econrate' with the median of the column in the 'bowl' DataFrame
bowl['econrate'].fillna(bowl['econrate'].median(), inplace=True)
```

```python
from pycaret.regression import *
df_sampled = bowl.sample(frac=0.5, random_state=42).reset_index(drop=True)

regression_setup = setup(data=df_sampled, target='dr11score', session_id=123, html = False)
best = compare_models()
best
```

```python
evaluate_model(best)
```
