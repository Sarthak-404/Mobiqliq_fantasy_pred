```python
import pandas as pd

def clean_data_team_1(url):
    # Load the data from the URL
    tables = pd.read_html(url)

    # Function to clean batting data
    def clean_batting(df):
        # Drop unnecessary columns
        df = df.drop(['Unnamed: 8', 'Unnamed: 9', 'Unnamed: 1', 'M'], axis=1)

        # Drop rows with NaN in the 'R' column as those rows are irrelevant (like Extras, Total)
        df = df.dropna(subset=['R'])

        # Drop rows where 'BATTING' contains 'Extras', 'TOTAL'
        df = df[~df['BATTING'].isin(['Extras', 'TOTAL'])]

        # Rename the 'BATTING' column to 'PLAYERS'
        df = df.rename(columns={'BATTING': 'PLAYERS'})

        # Reset the index to start from 0
        df = df.reset_index(drop=True)

        # Define a function to check if all values in a row are the same
        def all_columns_same(row):
            return all(row == row.iloc[0])

        # Apply this function along the rows and filter out those rows where the condition is True
        df = df[~df.apply(all_columns_same, axis=1)]

        # Reset the index if needed
        df = df.reset_index(drop=True)

        return df

    # Function to clean bowling data
    def clean_bowling(df):
        # Drop rows where 'W' column contains non-numeric data (i.e., drop commentary rows)
        df = df[pd.to_numeric(df['W'], errors='coerce').notna()]

        # Optionally, convert all numeric columns to their proper types
        numeric_columns = ['O', 'M', 'R', 'W', 'ECON', '0s', '4s', '6s', 'WD', 'NB']
        df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

        # Reset index after filtering
        df = df.reset_index(drop=True)

        return df

    # Clean batting data (assuming tables[0] is the batting table)
    cleaned_batting = clean_batting(tables[0])

    # Clean bowling data (assuming tables[3] is the bowling table)
    cleaned_bowling = clean_bowling(tables[3].rename(columns={'BOWLING': 'PLAYERS'}))

    # Concatenate the cleaned batting and bowling data
    combined_df = pd.concat([cleaned_batting, cleaned_bowling], axis=0, ignore_index=True)

    return combined_df

# Usage
url = 'https://www.espncricinfo.com/series/caribbean-premier-league-2023-1369538/trinbago-knight-riders-vs-guyana-amazon-warriors-final-1369591/full-scorecard'
team_1 = clean_data_team_1(url)

# Display the resulting DataFrame
team_1
```

# **Combined Code**

```python
import pandas as pd

class CricketDataCleaner:
    def __init__(self, url):
        self.url = url
        self.tables = pd.read_html(url)

    def clean_batting(self, df):
        df = df.drop(['Unnamed: 8', 'Unnamed: 9', 'Unnamed: 1', 'M'], axis=1)
        df = df.dropna(subset=['R'])
        df = df[~df['BATTING'].isin(['Extras', 'TOTAL'])]
        df = df.rename(columns={'BATTING': 'PLAYERS'})
        df['PLAYERS'] = df['PLAYERS'].str.replace(r'(\s*\(c\))|(\s*†)', '', regex=True)
        df = df.reset_index(drop=True)

        def all_columns_same(row):
            return all(row == row.iloc[0])

        df = df[~df.apply(all_columns_same, axis=1)]
        df = df.reset_index(drop=True)

        return df

    def clean_bowling(self, df):
        df = df.rename(columns={'BOWLING': 'PLAYERS'})
        df['PLAYERS'] = df['PLAYERS'].str.replace(r'(\s*\(c\))|(\s*†)', '', regex=True)
        df = df[pd.to_numeric(df['W'], errors='coerce').notna()]
        numeric_columns = ['O', 'M', 'R', 'W', 'ECON', '0s', '4s', '6s', 'WD', 'NB']
        df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')
        df = df.reset_index(drop=True)

        return df

    def clean_team_data(self, batting_index, bowling_index):
        cleaned_batting = self.clean_batting(self.tables[batting_index])
        cleaned_bowling = self.clean_bowling(self.tables[bowling_index])
        combined_df = pd.concat([cleaned_batting, cleaned_bowling], axis=0, ignore_index=True)

        return combined_df
```

```python
url = 'https://www.espncricinfo.com/series/caribbean-premier-league-2023-1369538/trinbago-knight-riders-vs-guyana-amazon-warriors-final-1369591/full-scorecard'
cleaner = CricketDataCleaner(url)

team_1_data = cleaner.clean_team_data(batting_index=0, bowling_index=3)
team_2_data = cleaner.clean_team_data(batting_index=2, bowling_index=1)
```
