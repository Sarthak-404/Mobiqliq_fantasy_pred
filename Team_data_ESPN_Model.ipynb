```python
sl_combine = pd.concat([sl_bat_odi_1, sl_bat_odi_2, sl_bat_odi_3, sl_bat_odi_4, sl_ball_odi_1, sl_ball_odi_2, sl_ball_odi_3, sl_ball_odi_4], ignore_index=True)

sl_combine['PLAYERS'] = sl_combine['PLAYERS'].str.replace(r'(\s*\(c\))|(\s*†)', '', regex=True)

sl_combine[['R', 'B', 'M', '4s', '6s', 'SR','O','W','ECON','WD','NB']] = sl_combine[['R', 'B', 'M', '4s', '6s', 'SR','O','W','ECON','WD','NB']].apply(pd.to_numeric, errors='coerce')

sl_grouped = sl_combine.groupby('PLAYERS').agg({
    'R': 'sum',
    'B': 'sum',
    'M': 'sum',
    '4s': 'sum',
    '6s': 'sum',
    'SR': 'mean',
    'O': 'sum',
    'W': 'sum',
    'ECON': 'mean',
    'WD': 'sum',
    'NB': 'sum'
}).reset_index()

appearances = sl_combine['PLAYERS'].value_counts().reindex(sl_grouped['PLAYERS']).fillna(1).values

sl_grouped['avg_R'] = sl_grouped['R'] / appearances
sl_grouped['avg_B'] = sl_grouped['B'] / appearances
sl_grouped['avg_M'] = sl_grouped['M'] / appearances
sl_grouped['avg_4s'] = sl_grouped['4s'] / appearances
sl_grouped['avg_6s'] = sl_grouped['6s'] / appearances
sl_grouped['avg_SR'] = sl_grouped['SR']
sl_grouped['avg_O'] = sl_grouped['O'] / appearances
sl_grouped['avg_W'] = sl_grouped['W'] / appearances
sl_grouped['avg_ECON'] = sl_grouped['ECON']
sl_grouped['avg_WD'] = sl_grouped['WD'] / appearances
sl_grouped['avg_NB'] = sl_grouped['NB'] / appearances

sl_grouped
```

```python
ir_combine = pd.concat([ir_bat_odi_1, ir_bat_odi_2, ir_bat_odi_3, ir_bat_odi_4, ir_ball_odi_1, ir_ball_odi_2, ir_ball_odi_3, ir_ball_odi_4], ignore_index=True)

ir_combine['PLAYERS'] = ir_combine['PLAYERS'].str.replace(r'(\s*\(c\))|(\s*†)', '', regex=True)

ir_combine[['R', 'B', 'M', '4s', '6s', 'SR','O','W','ECON','WD','NB']] = ir_combine[['R', 'B', 'M', '4s', '6s', 'SR','O','W','ECON','WD','NB']].apply(pd.to_numeric, errors='coerce')

ir_grouped = ir_combine.groupby('PLAYERS').agg({
    'R': 'sum',
    'B': 'sum',
    'M': lambda x: x.mode()[0] if not x.mode().empty else 0,
    '4s': lambda x: x.mode()[0] if not x.mode().empty else 0,
    '6s': lambda x: x.mode()[0] if not x.mode().empty else 0,
    'SR': 'mean',
    'O': 'sum',
    'W': lambda x: x.mode()[0] if not x.mode().empty else 0,
    'ECON': 'mean',
    'WD': 'sum',
    'NB': 'sum'
}).reset_index()

appearances = ir_combine['PLAYERS'].value_counts().reindex(ir_grouped['PLAYERS']).fillna(1).values

ir_grouped['avg_R'] = ir_grouped['R'] / appearances
ir_grouped['avg_B'] = ir_grouped['B'] / appearances
ir_grouped['mode_M'] = ir_grouped['M']
ir_grouped['mode_4s'] = ir_grouped['4s']
ir_grouped['mode_6s'] = ir_grouped['6s']
ir_grouped['avg_SR'] = ir_grouped['SR']
ir_grouped['avg_O'] = ir_grouped['O'] / appearances
ir_grouped['mode_W'] = ir_grouped['W']
ir_grouped['avg_ECON'] = ir_grouped['ECON']
ir_grouped['avg_WD'] = ir_grouped['WD'] / appearances
ir_grouped['avg_NB'] = ir_grouped['NB'] / appearances

ir_grouped
```

```python
numeric_columns = ['R', 'B', 'M', '4s', '6s', 'SR', 'O', 'W', 'ECON', 'WD',
       'NB', 'avg_R', 'avg_B', 'avg_M', 'avg_4s', 'avg_6s', 'avg_SR', 'avg_O',
       'avg_W', 'avg_ECON', 'avg_WD', 'avg_NB']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
def calculate_score(row):
  runs = row['avg_R']
  fours = row['avg_4s']
  sixes = row['avg_6s']
  sr = row['avg_SR']
  wickets = row['avg_W']
  wides = row['WD']
  econ = row['avg_ECON']
  no_balls = row['NB']
  overs = row['avg_O']
  fours = row['avg_4s']
  sixes = row['avg_6s']
  maidens = row['M']

  score = runs + fours + sixes + (wickets * 25) + (maidens * 4)

  if runs>=50:
    score+=4
  if wickets>=3:
    score+=2
  elif wickets>=4:
    score+=4
  elif wickets>=5:
    score+=8

  if sr<30:
    score-=6
  elif sr<40 and sr>=30:
    score-=4
  elif sr>=40 and sr<50:
    score-=2
  elif sr>=100 and sr<120:
    score+=2
  elif sr>=120 and sr<=140:
    score+=4
  elif sr>140:
    score+=6

  if econ<2.5:
    score+=6
  elif econ<3.5 and econ>=2.5:
    score+=4
  elif econ<4.5 and econ>=3.5:
    score+=2
  elif econ<8 and econ>=7:
    score-=2
  elif econ<9 and econ>=8:
    score-=4
  elif econ>=9:
    score-=6
  return score
```

```python
df['fantasy score'] = df.apply(calculate_score, axis=1)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
df[numeric_columns] = imputer.fit_transform(df[numeric_columns])
```

```python
df = df.drop(['B','O','WD','avg_WD','avg_NB','avg_O','NB','avg_B'],axis=1)  # 'R' was in this previously
```

```python
X = df.drop(['fantasy score','PLAYERS'],axis=1)
y = df['fantasy score']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size = 0.8, random_state = 42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
model_0 = RandomForestRegressor(n_estimators=100, random_state=42)
model_0.fit(X_train,y_train)
feature_importance = model_0.feature_importances_
feature_importance
```

```python
import matplotlib.pyplot as plt
features = X_train.columns
feature_importance_df = pd.DataFrame({
    'Feature': features,
    'Importance': feature_importance
})

feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 8))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'], color='skyblue')
plt.xlabel('Feature Importance')
plt.ylabel('Features')
plt.title('Feature Importance from RandomForestRegressor')
plt.gca().invert_yaxis()
plt.show()
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error, r2_score
xg_reg = xgb.XGBRegressor(objective='reg:squarederror', colsample_bytree=0.3, learning_rate=0.1,
                          max_depth=5, alpha=10, n_estimators=100, random_state=42)

xg_reg.fit(X_train, y_train)
```

```python
y_pred = xg_reg.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

```python
xgb.plot_importance(xg_reg, importance_type='weight')
plt.show()
```
